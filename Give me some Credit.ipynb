```python
x_train = train.drop(["SeriousDlqin2yrs","Unnamed: 0"],1)
x_train.head()
```

```python
y_train = train["SeriousDlqin2yrs"]
y_train.head()
```

```python
x_test =  test.drop(["SeriousDlqin2yrs","Unnamed: 0"],1)
```

```python
x_test.head()
```

```python
from xgboost import XGBClassifier 
xgb = XGBClassifier(nthread = 4,colsample_bytree=0.8,subsample=0.7, max_depth=4,n_estimators=440,learning_rate=0.025)
```

```python
xgb.fit(x_train,y_train)
result1 = xgb.predict_proba(x_test)
result1
```

1. 하이퍼파라미터 설정 중요 nthread ; n_jobs와 동일,lgbm은 필요없음
2. lgbm,xgboost : 나무 기본값 100개
3. max_depth : tree가 학습할 수 있는 양 / max_depth가 커지면 데이터셋에 대해 가지치기를 더많이 해서 세분화하여 많은 학습/ 복잡하게 학습; 데이터가 복잡하면 max_depth 올리기/기본값 3 -하나씩 올리면서(내리면서) 성능 비교 ;  
4. n_estimators도 중요; 100이 기본값/ 올렸을때 좋은지 낮췄을 때 좋은지 실험해보기 이 대회의 경우 max_depth크게, n_estimators 크게...
5. 최적값을 찾을 때 속도 - 1. 주변을 둘러보면서 계속 확인; 속도 느림(learning_rate가 낮음) -기본적으로 learning_rate가 낮은것이 더 좋음;최적점에 훨씬 가까이 잘 다가갈 수 있음/ but 시간이 오래걸림;학습하다가 멈출수도 있음; global optimum을 찾아가야 하는데 local minimum에 안착../ 최적점을 잘 찾아가지 못하고 페널티를 얻을 수 도 있음; 
  2. 속도있게 진행하다가 가끔씩 주변을 보는(learning_rate가 큼) : 빠르게 최적점까지 갈 수 있음/ but step size가 커서(학습속도가 커서) 최적점을 지나칠 수 있음; 모델의 점수가 잘 나오지 않을 수 있음; 최적점으로 수렴하지 못하고 발산할 수 있음 ; 
 5-1. 기본값 0.1 거의 대부분 작을때 성능 향상 (0.01보다 작으면 성능 하락) 최적: 0.01 ~ 0.025
 5-2.  러닝레이트 튜닝하기 : 러닝레이트를 낮춘다 -> 각 나무가 학습하는 양을 줄인다. -> 나무의 개수를 더 많이 만들어줘야함/ 0.1->0.025 4배만큼 나무의 효율성이 낮아짐; 나무의 갯수를 4배 늘려줘야함; 

```python
from lightgbm import LGBMClassifier 
lgb = LGBMClassifier(colsample_bytree=0.8,subsample=0.7, num_leaves= 25,n_estimators=450, learning_rate=0.02)
#num_leaves는 31이 기본, 
```

# 모델 성능 높이기 -> 하이퍼파라미터 조정
### colsample_bytree:기본값 1 (0~1사이의 값을가짐) 변수를 해당비율만큼 가져와서 학습하기 (랜덤하게 뽑기)
->트리모델은 학습에 도움이되는 feature에만 편향하는 경향이 있음; 일부 칼럼만 집중적으로 학습; 과적합 유발(특정 칼럼에 대한) train과 test의 분포가 무조건 똑같지 않기 때문에 train에서 학습을 잘했던 게 test에서는 아닐 수 있음; 일정 칼럼만 독점적으로 학습하는 게 좋지 않음
-> 칼럼을 샘플링해서 뽑아서 과적합 방지; 몇몇 나무에서는 중요하다고 생각되지 않은(중요한 칼럼이 빠진) feature들이 학습에 이용됨; 각 나무들의 다양성 확보 (앙상블 효과 ) 새로운 패턴의 데이터에 대해서도 처리할 수 있음; 모든 대회에서 중요한 하이퍼파라미터; 
-> (주의) : 다른 하이퍼파라미터와 같이 , 최종적인 모델의 점수 향상 ; 0.5~0.8 칼럼이 많으면 0.3~0.4(숫자는 크게 중요하지 않음)
